## 05 - Exibindo Página Web

Agora que já sabemos coletar as palavras mais frequentes, vamos montar uma página web capaz de exibir os dados coletados. Para que o resultado seja satisfatório precisamos de uma base de dados com um número grande de tweets. Como o tempo do minicurso não permite a gente coletar essa base, vamos utilizar um conjunto de dados coletados anteriormente.

Os dados coletados consistem de tweets coletados de 03/04 à 01/05 (não contínuos) que tinham as hashtags #teambatman ou #teamsuperman. Foram coletados 653 tweets de 601 usuários distintos.

Para exibir a página vamos utilizar um framework web para Python chamado Flask (http://flask.pocoo.org/) e vamos armazernar os dados em uma base sqllite.

O propósito da página é apenas exibir as informações coletadas de forma que possamos entender melhor aquele conjunto de dados.
